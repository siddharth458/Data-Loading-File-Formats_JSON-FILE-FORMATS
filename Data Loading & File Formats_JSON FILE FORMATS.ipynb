```python
import sqlite3
conn = sqlite3.connect("employee_db")
```

```python
df = pd.read_sql("select*from emp;",conn)
df.head()
```

```python
df = pd.read_sql("""SELECT e.ename,e.job,e.deptno,d.dname
                      FROM emp as e 
                      INNER JOIN dept as d 
                      ON d.deptno = e.deptno;"""
                          , conn)
df.head()
```

```python
df = pd.read_sql("""SELECT e.ename,e.job,e.deptno,d.dname
                      FROM emp as e 
                      LEFT JOIN dept as d 
                      ON d.deptno = e.deptno;"""
                          , conn)
df.head()
conn.close()
```
